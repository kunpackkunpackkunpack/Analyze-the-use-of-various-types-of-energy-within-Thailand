```python
x1 = np.array(ytdfu_all['Year'])
x2 = np.array(ytd_all['Year'])
y1 = np.array(ytdfu_all['Total'])
y2 = np.array(ytd_all['Total'])

model1 = LinearRegression(fit_intercept=True)
model2 = LinearRegression(fit_intercept=True)

model1.fit(x1[:, np.newaxis], y1)
xfit1 = np.linspace(2002, 2021, 2)
yfit1 = model1.predict(xfit1[:, np.newaxis])


model2.fit(x2[:, np.newaxis], y2)
xfit2 = np.linspace(2002, 2021, 2)
yfit2 = model2.predict(xfit2[:, np.newaxis])

plt.plot(x1, y1)
plt.plot(x2, y2)

plt.plot(xfit1, yfit1);
plt.plot(xfit2, yfit2);
plt.xlabel('Year', fontsize=15)
plt.ylabel('Average unit', fontsize=15)
```
